```python
import pandas as pd

data = {
    "Sensor_ID": [101, 102, 103, 103, 104, 105],
    "Temperature": [28.5, None, 30.2, 30.2, 27.8, 31.0],
    "Humidity": [65, 60, None, None, 70, 55],
    "Location": ["Room1", "Room2", "Room3", "Room3", "Room1", "Room2"]
}

df = pd.DataFrame(data)

df
```

```python
df.to_csv("raw_sensor_data.csv", index=False)

print("Raw Dataset Saved Successfully")
```

```python
df = pd.read_csv("raw_sensor_data.csv")

print("Original Dataset")
df
```

```python
print("Missing Values in Each Column")

df.isnull().sum()
```

```python
df["Temperature"] = df["Temperature"].fillna(df["Temperature"].mean())

df["Humidity"] = df["Humidity"].fillna(df["Humidity"].mean())

print("Missing Values Filled Successfully")

df
```

```python
df = df.drop_duplicates()

print("Duplicate Rows Removed Successfully")

df
```

```python
print("Dataset Shape:")

df.shape
```

```python
df["Temperature_Status"] = df["Temperature"].apply(
    lambda x: "High" if x > 30 else "Normal"
)

print("Feature Extraction Completed")

df
```

```python
df.to_csv("cleaned_sensor_data.csv", index=False)

print("Cleaned Dataset Saved Successfully")
```
